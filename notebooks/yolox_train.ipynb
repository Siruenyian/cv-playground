````python
import json

p = "/app/datasets/rps/annotations/train.json"
out = "/app/datasets/rps/annotations/train_fixed.json"

with open(p) as f:
    data = json.load(f)

# Keep categories with id != 0 and sort by original id
cats = [c for c in data["categories"] if c["id"] != 0]
cats.sort(key=lambda c: c["id"])

# Build mapping old_id -> new_id (0..)
mapping = {c["id"]: i for i, c in enumerate(cats)}

# Update categories ids & names
new_cats = []
for old_id, new_id in mapping.items():
    c = next(c for c in cats if c["id"] == old_id)
    c["id"] = new_id
    new_cats.append(c)

# Update annotations
for ann in data["annotations"]:
    old = ann["category_id"]
    if old == 0:
        ann["category_id"] = -1  # or decide how to handle (remove if unexpected)
    else:
        ann["category_id"] = mapping[old]

data["categories"] = new_cats
# Optionally filter out annotations with category_id == -1
data["annotations"] = [a for a in data["annotations"] if a["category_id"] != -1]

with open(out, "w") as f:
    json.dump(data, f, indent=2)
print("Wrote", out)
````

## 3. Prepare Dataset Structure

YOLOX expects:
```
datasets/rps/
  annotations/train.json
  annotations/val.json
  train2017/  (images)
  val2017/    (images)
```
Symlinks map the Roboflow download into this layout.

````python
import os

DATASET_SRC = "/app/notebooks/rock-paper-scissors-14"
DATASET_DST = "/app/datasets/rps"

os.makedirs(f"{DATASET_DST}/annotations", exist_ok=True)

for src_split, dst_name in [("train", "train.json"), ("valid", "val.json")]:
    src = f"{DATASET_SRC}/{src_split}/_annotations.coco.json"
    dst = f"{DATASET_DST}/annotations/{dst_name}"
    if not os.path.lexists(dst):
        os.symlink(src, dst)
    print(f"{dst} -> {src}")

for src_split, dst_name in [("train", "train2017"), ("valid", "val2017")]:
    src = f"{DATASET_SRC}/{src_split}"
    dst = f"{DATASET_DST}/{dst_name}"
    if not os.path.lexists(dst):
        os.symlink(src, dst)
    print(f"{dst} -> {src}")

print("\nDataset structure ready!")
````
